```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from pathlib import Path
import glob
import altair as alt
import numpy as np
import nltk


```

```python
directory_path = "/content/drive/MyDrive/Colab Notebooks/Python Project"
text_files = glob.glob(f"{directory_path}/*.txt")
```

```python
text_files
```

```python
text_titles = [Path(text).stem for text in text_files]
```

```python
tfidf_vectorizer = TfidfVectorizer(input='filename', stop_words='english')
tfidf_vector = tfidf_vectorizer.fit_transform(text_files)

```

```python
tfidf_df = pd.DataFrame(tfidf_vector.toarray(), index=text_titles, columns=tfidf_vectorizer.get_feature_names_out())
```

```python
tfidf_df
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
import glob

# Download NLTK data
nltk.download('wordnet')

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# List of file paths
file_paths = glob.glob("/content/drive/MyDrive/Colab Notebooks/Python Project/*.txt")

# Loop through each file
for file_path in file_paths:
    with open(file_path, 'r', encoding='utf-8') as file:
        # Read the content of the file
        text = file.read()

        # Tokenize the text into words
        words = text.split()

        # Lemmatize each word
        lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

        # Join the lemmatized words back into a sentence
        lemmatized_text = ' '.join(lemmatized_words)

    # Save the lemmatized text back to the file
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(lemmatized_text)
```
